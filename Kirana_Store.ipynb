# **Import Library**


```python
import datetime
import pandas as pd
```

# **Import Data**

```python
#Sample inventory data for a Kirana Store
sales_data = {
    'Item': ['Rice(kg)','Atta(kg)','Dal(kg)','Sugar(kg)','Oil(liter)','Spices(pkt)','Tea(pkt)','Biscuit(pkt)','Pen','Pencile','Coffee(pkt)'],
    'Quantity': [100, 80, 50, 60, 100, 100, 50, 120, 50, 60, 90],
    'Price':[60, 40, 50, 45, 150, 30, 80, 20, 10, 5, 20]
}
inventory = pd.DataFrame(sales_data)
```

```python
# function to update inventory after a sale
def update_inventory(item, quantity):
   global daily_sales_list
   daily_sales_list = []
   if item in inventory['Item'].values:
    index = inventory[inventory['Item'] == item].index[0]
    if inventory.loc[index, 'Quantity'] >= quantity:
      inventory.loc[index, 'Quantity'] -= quantity
      daily_sales_list.append({'Item': item, "Quantity": quantity})
      print(f"{quantity} {item} is sold.")
    else:
      print(f"Not enough {item} in stock.")
   else:
    print(f"{item} not found in inventory.")
```

```python
#function to generate a bill
def generate_bill(items,quantities,date):
  total_bill = 0
  print(f"--- Bill ({date.strftime('%Y-%m-%d') if date else 'N/A'})---")
  for item, quantity in zip(items, quantities):
    if item in inventory['Item'].values:
      index = inventory[inventory['Item'] == item].index[0]
      price = inventory.loc[index, 'Price']
      cost = price * quantity
      print(f"{item} x {quantity} = {cost:.2f}")
      total_bill += cost
    else:
      print(f"{item} not found in inventory.")
  print(f"Total: {total_bill:.2f}")
```

```python
#function to update price
def update_price(items, new_prices):
  for item, new_price in zip(items, new_prices):
    if item in inventory['Item'].values:
      index = inventory[inventory['Item'] == item].index[0]
      inventory.loc[index, 'Price'] =new_price
      print(f"price of {item} updated to {new_price:2f}")
    else:
      print(f"{item} not found in inventory")
```

```python
#Example usage
update_inventory('Coffee(pkt)', 10)
update_inventory('Pen', 10)
update_inventory('Sugar(kg)',30)
update_inventory('Atta(kg)',20)
```

```python
#Bill
today = datetime.date.today()
generate_bill(['Coffee(pkt)','Pen','Sugar(kg)','Atta(kg)'],[10,10,30,20], date=today)
```

```python
#check for low stock
low_stock_items = inventory[inventory['Quantity'] < 20]['Item'].tolist()
if low_stock_items:
  print("\nLow stock alert:")
  for item in low_stock_items:
    print(f"{item} is running low.")
else:
  print("\nAll items are sufficiently stocked.")
```
